```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
company=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
x=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
#x = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
```

### Plot

```python
fig = plt.figure(figsize=(15,3)) # Size of grap(Horizontal, Vertical)

plt.plot(company,x)
plt.xlabel('Companies')
plt.ylabel('Measure')
plt.title('Line Plot')
```

```python
#Creating histogram
plt.style.use('ggplot')
plt.hist(x)

# Show plot
plt.show()
```

### Boxplot

```python
# Box Plot to find outliars
sns.boxplot(x)
```
